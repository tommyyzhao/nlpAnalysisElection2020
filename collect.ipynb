```
import os
import json
import csv
from tqdm import tqdm
```

```
# get the path for the raw data
raw_path = os.path.join('drive','My Drive','Colab Notebooks','Raw')

# obtain all the files in the Raw folder
files = os.listdir(raw_path)

# define interim folder path
interim_dir = os.path.join('drive','My Drive','Colab Notebooks', 'Interim')

# make interim directory
if not os.path.exists(interim_dir):
  os.makedirs(interim_dir)

collected_path = os.path.join(interim_dir, 'collected.jsonl')
with open(collected_path, 'w') as collected:
  for file in tqdm(files):
    total_path = os.path.join(raw_path, file)
    
    concat = []
    with open(total_path, 'r') as f:
      for line in f.readlines():
        j_line = json.loads(line)
        sorted_j = {i: j_line[i] for i in sorted(list(j_line.keys()))}
        json.dump(sorted_j, collected)
        collected.write('\n')

# check if file stored
if os.path.exists(collected_path):
  print('\ncollected.jsonl stored in Interim.')
```
